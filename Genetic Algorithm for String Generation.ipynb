```python
import random
import numpy as np

chrs = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789 ~!@#$%^&*()_-`{[()}]:;.,<>?/\|+=*'
target = 'Hello, World!'
n = 70

population = []
for i in range(n):
    population.append(''.join(random.choices(chrs, k=len(target))))
```

```python
def Fitness(a, b):
    fit = 0
    for x, y in zip(a, b):
        if x != y:
            fit+=1
    return fit

def mutate():
    return random.choice(chrs)

def reproduce(parent1, parent2):
    child = ''
    for x,y in zip(parent1, parent2):
        prob = random.random()
        if prob < 0.45:
            child += child.join(x)
        elif prob < 0.9:
            child += child.join(y)
        else:
            child += child.join(mutate())
    return child
```

```python
def GeneticAlgorithm(population, target, chrs):

    found = False
    gen = 1
    n = len(population)
    while not found:
        population = sorted(population, key = lambda x: Fitness(x, target))
        if Fitness(population[0], target) <= 0 :
            found = True
            break
        new_generation = []

        s = int((10*n)/100)
        new_generation.extend(population[:s])

        s = int((90*n)/100)
        for _ in range(s):
            parent1 = random.choice(population[:n//3]) # randoming choicing chromosome from 1/3 of sorted array
            parent2 = random.choice(population[:n//3]) # to reduce overhead :D (isn't neccessary )
            child = reproduce(parent1, parent2)
            new_generation.append(child)

        print("Generation:   ", gen, end='\t')
        print("String: ", population[0], end='\t')
        print("Fitness: ", Fitness(population[0], target))

        population = new_generation
        gen +=1

    print("Generation:   ", gen, end='\t')
    print("String: ", population[0], end='\t')
    print("Fitness: ", Fitness(population[0], target))
```

```python
GeneticAlgorithm(population, target, chrs)
```
