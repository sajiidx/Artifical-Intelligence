```python
#Question 02
def astar(graph, start, goal, gn, hn, open_q, close_q=[], cost=0):
    print(start, end='-->')
    close_q.append(start)
    open_q.remove(start)
    if start == goal:
        return cost

    mini = float('inf')
    node = ''
    temp = cost
    for i in range(len(graph[start])):
        open_q.append(graph[start][i])
        
    for i in range(len(open_q)):
        if str(start + '-' + open_q[i]) in gn.keys():
            fn = cost + (gn[start + '-' + open_q[i]] + hn[open_q[i]])
            if mini > fn:
                mini, node = fn, open_q[i]
                temp = gn[start + '-' + open_q[i]]
    return astar(graph, node, goal, gn, hn, open_q, close_q, cost+temp)
graph = { 'S': ['A', 'G'], 'A': ['B', 'C'], 'B': ['D'], 'C': ['D','G'], 'D': ['G'], 'G': [] }
hn = {'A': 3, 'B': 4, 'C':2, 'D':6, 'S': 5,'G':0}
gn = {'S-A': 1, 'S-G': 10, 'A-B': 2,'A-C': 1, 'B-D':5, 'C-D':3,'C-G':4,'D-G':6}
print(astar(graph, 'S', 'G', gn, hn, ['S']))
```

```python
#Question 04
import copy as cp
import numpy as np
def find(node, inmaze):
    for row in range(len(inmaze)):
        for col in range(len(inmaze[row])):
            if inmaze[row][col] == node:
                return (row, col)
                
def ChangeState(state, dash, move):
    xtemp = cp.deepcopy(state)
    i, j = dash
    x, y = move
    xtemp[i][j], xtemp[x][y] = xtemp[x][y], xtemp[i][j]
    return xtemp

def h(state, goal):
    h = 1
    for i in range(len(state)):
        for j in range(len(state[i])):
            if state[i][j] != goal[i][j]:
                h+=1
    return h

def arraytostring(state):
    ss = ''
    for row in state:
        for num in row:
            ss += str(num)
            
    return ss
                       

def sovlePuzzleAstar(puzzle, goal, dash, gn, open_q, close_q=[]):
    print(np.array(puzzle))
    
    if np.array_equal(puzzle, goal):
        return 0
    
    close_q.append(cp.deepcopy(puzzle))
    open_q.remove(cp.deepcopy(puzzle))
    
    
    
    mini = float('inf')
    i, j = dash
    r, c = (3, 3)
    node = []
    k = gn[arraytostring(cp.deepcopy(puzzle))]
    
    
    if j+1 < c:
        dummy = cp.deepcopy((ChangeState(cp.deepcopy(puzzle), dash, (i, j+1))))
        if not dummy in open_q and not dummy in close_q:
            open_q.append(cp.deepcopy(dummy))
            gn[arraytostring(cp.deepcopy(dummy))] = k+1
        
    if i+1 < r:
        dummy = cp.deepcopy(ChangeState(cp.deepcopy(puzzle), dash, (i+1, j)))
        if not dummy in open_q and not dummy in close_q:
            open_q.append(cp.deepcopy(dummy))
            gn[arraytostring(cp.deepcopy(dummy))] = k+1
        
    if j-1 >= 0:
        dummy = cp.deepcopy(ChangeState(cp.deepcopy(puzzle), dash, (i, j-1)))
        if not dummy in open_q and not dummy in close_q:
            open_q.append(cp.deepcopy(dummy))
            gn[arraytostring(cp.deepcopy(dummy))] = k+1
        
    if i-1 >= 0:
        dummy = cp.deepcopy(ChangeState(cp.deepcopy(puzzle), dash, (i-1, j)))
        if not dummy in open_q and not dummy in close_q:
            open_q.append(cp.deepcopy(dummy))
            gn[arraytostring(cp.deepcopy(dummy))] = k+1
    for game in open_q:
        fn = h(cp.deepcopy(game), goal) + gn[arraytostring(cp.deepcopy(game))]
        if mini > fn:
            mini, node = fn, cp.deepcopy(game)
            
    return mini + sovlePuzzleAstar(node, goal, find(0, node), gn, open_q, close_q)
    
initial = [
    [1, 2, 3],
    [0, 4, 6],
    [7, 5, 8]
]
goal = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 0]
]
sovlePuzzleAstar(initial, goal, find(0, initial), {arraytostring(initial): 0}, [initial])
```
