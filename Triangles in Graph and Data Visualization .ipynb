```python
class Tree:
    class Node:
        def __init__(self,data,left=None,right=None):
            self.data = data
            self.left = left
            self.right = right
            
        def printNode(self):
            print(self.data)
    
    def __init__(self,data = None):
        self.root = self.Node(data)
        
    def insert(self,root,data):
        if root is None:
            return self.Node(data)
        else:
            if root.data == data:
                return root
            elif root.data < data:
                root.right = self.insert(root.right,data)
            else:
                root.left = self.insert(root.left, data)
        return root
    
    def insrt(self,data):
        self.root = self.insert(self.root, data)
    
    def printTree(self, root):
        if root:
            self.printTree(root.left)
            root.printNode()
            self.printTree(root.right)
#----------------------------Q2------------------------------
            
    def size(self, root):
        if root:
            res = 1
            res = res + self.size(root.left)
            res = res + self.size(root.right)
            return res
        return 0
    
#----------------------------Q3------------------------------

    def mirror(self, root):
        if root:
            root.left = self.mirror(root.left)
            root.right = self.mirror(root.right)
            
            temp = root.left
            root.left = root.right
            root.right = temp
            
            return root
        
bst = Tree(1)
bst.insrt(8)
bst.insrt(3)
bst.insrt(6)
bst.insrt(7)
print('Size: ', bst.size(bst.root))
print('Inorder: ')
bst.printTree(bst.root)
print('Mirror: ')
bst.mirror(bst.root)
bst.printTree(bst.root)
#----------------------------------------------------------
```
